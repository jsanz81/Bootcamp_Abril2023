````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
plt.plot([0,1,2,3,4,5])
````

## Funciones que podemos utilizar
* text(posicion_x,posicion_y,text)

````python
plt.plot([0,1,2,3,4,5])
plt.text(2,3,'Hola')
````

* Etiquetas
    * Eje de la x ```xlabel```
    * Eje de la y ```ylabel```

````python
plt.plot([0,1,2,3,4,5])
plt.xlabel('Hola')
plt.ylabel('Adios')
````

* Titulo ```title```

````python
plt.plot([0,1,2,3,4,5])
plt.title('Bisectriz')
````

* Hacer la cuadricula ```grid```

````python
plt.plot([0,1,2,3,4,5])
plt.grid()
````

````python

x=np.linspace(-10,10,1000)
y=x**2
````

````python
y
````

````python
plt.plot(x,y,color="yellow")
````

````python
plt.plot(x,y,color="green")
````

````python
plt.plot(x,y,color="green",label="Parabola")
plt.grid()
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title("Parabola")
plt.legend()
````

````python
z=x**3
````

````python
plt.plot(x,y,color="green",label="Parabola")
plt.plot(x,z,color="red",label="Cubico")
plt.grid()
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title("Combinacion")
plt.legend()
````

````python
plt.plot(x,y,color="green",label="Parabola",marker='^')
plt.plot(x,z,color="red",label="Cubico",marker='o')
plt.grid()
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title("Combinacion")
plt.legend()
````

````python
x=np.array([1,2,3,4,5,6,7,8])
y=np.random.randint(1,100,size=len(x))
````

````python
y
````

````python
### Grafico de dispersin
plt.scatter(x,y,color="cyan")
plt.title('Dispersin')
plt.xlabel('Temperatura')
plt.ylabel('Velocidad aire');
````

````python
### Grafico de barras
````

````python
partidos=["Partido1","Partido2","Partido3","Partido4"]
votos=[10,20,30,40]
explode=(0.1,0,0,0.2) ## Para sacar algn sector
plt.pie(votos,labels=partidos,explode=explode,autopct="%1.1f",colors=["yellow","pink","cyan","brown"]);
````
