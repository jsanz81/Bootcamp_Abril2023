```python
from sklearn import datasets
```

```python
datos=datasets.load_breast_cancer()
```

```python
import pandas as pd
import numpy as np 
```

```python
print(datos["DESCR"])
```

```python
df=pd.DataFrame(datos["data"],columns=datos["feature_names"])
```

```python
df["target"]=datos.target
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test = train_test_split(df,test_size=0.25)
```

```python
df_train.shape
```

```python
regresion_logistica=LogisticRegression()
```

```python
regresion_logistica.fit(df_train.drop(columns="target"),df_train["target"])
```

```python
regresion_logistica.coef_
```

```python
y_pred=regresion_logistica.predict(df_test.drop(columns="target"))
```

```python
import matplotlib.pyplot as plt
```

```python
df_test["y_pred"]=y_pred
```

```python
plt.scatter(df_test["y_pred"],df_test["target"])
```

```python
pd.crosstab(df_test["y_pred"],df_test["target"])
```

```python
df_test["proba"]=regresion_logistica.predict_proba(df_test.drop(columns=["target","y_pred"]))[:,1]
```

```python
df_test[df_test["target"]!=df_test["y_pred"]]
```

```python
predicciones=regresion_logistica.predict_proba(df_test.drop(columns=["target","y_pred","proba"]))
```
